```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import pandas as pd
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
digits.data.shape
```

```python
digits.target
```

```python
digits.data[:1]
```

```python
data_X=digits.data[:600]
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=3, random_state=0)
```

```python
tsne_obj = tsne.fit_transform(data_X)
```

```python
tsne_df = pd.DataFrame({'X':tsne_obj[:,0],
                        'Y':tsne_obj[:,1],
                        'Z':tsne_obj[:,2]})
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10)
labels= kmeans.fit(tsne_df)
```

```python
tsne_df['label'] = labels.labels_
```

```python
tsne_df
```

```python
df1 = tsne_df.loc[tsne_df['label'] == 1]
df2 = tsne_df.loc[tsne_df['label'] == 4]
```

```python
df2
```

```python
# sns.scatterplot(x="X",y="Y",data=tsne_df[['X','Y']])
# import plotly.express as px
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (10, 7)) 
ax = plt.axes(projection ="3d") 

# ax.scatter(xs, ys, zs, c='r', marker='o', label='r')
# ax.scatter(xt, yt, zt, c='b', marker='^', label='b')

ax.scatter(df1['X'], df1['Y'], df1['Z'], color = "green", label='1') 
ax.scatter(df2['X'], df2['Y'], df2['Z'], color = "red", label='4')

ax.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10)
labels= kmeans.fit(tsne_df)
```

```python
labels.labels_[10]
```

```python
tsne_df['label'] = labels.labels_
```

```python
sns.scatterplot(x="X", y="Y",
              hue="label",
              palette=['purple','red','orange','brown','blue',
                       'dodgerblue','green','lightgreen','darkcyan', 'black'],
              legend='full',
              data=tsne_df[['X','Y','label']]);
```

```python
# kernelized k means
from sklearn.cluster import SpectralClustering
model = SpectralClustering(n_clusters=10, affinity='nearest_neighbors',
                           assign_labels='kmeans')
labels2 = model.fit(tsne_df[['X','Y']])
```

```python
tsne_df['label2'] = labels2.labels_
```

```python
sns.scatterplot(x="X", y="Y",
              hue="label2",
              palette=['purple','red','orange','brown','blue',
                       'dodgerblue','green','lightgreen','darkcyan', 'black'],
              legend='full',
              data=tsne_df[['X','Y','label2']]);
```
